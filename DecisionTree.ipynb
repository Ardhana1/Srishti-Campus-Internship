```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_test=sc.fit_transform(x_test)
x_train=sc.transform(x_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
clf2 = DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
clf2.fit(x_train, y_train)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
print(clf2.predict(sc.transform([[51,87000]])))
```

```python
y_pred = classifier.predict(x_test)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
y_pred2 = clf2.predict(x_test)
#print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm2 = confusion_matrix(y_test, y_pred2)
print(cm2)
accuracy_score(y_test, y_pred2)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
        classifier,
        x_train,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )
# Scatter plot
plt.scatter(x_train[:, 0], x_train[:, 1], 
            c=y_train, 
            s=20, edgecolors="k")
plt.title('Decision Tree-Classifier(Training set)')
plt.show()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
# Plot Decision Boundary
DecisionBoundaryDisplay.from_estimator(
        classifier,
        x_test,
        response_method="predict",
        cmap=plt.cm.Spectral,
        alpha=0.8,
        xlabel=dataset.columns[0],
        ylabel=dataset.columns[1],
    )
# Scatter plot
plt.scatter(x_test[:, 0], x_test[:, 1], 
            c=y_test, 
            s=20, edgecolors="k")
plt.title('Decision Tree-Classifier(Testing set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
def plot_decision_tree(clf_object, feature_names, class_names):
    plt.figure(figsize=(15, 10))
    tree.plot_tree(clf_object, 
                   filled=True, 
                   feature_names=feature_names, 
                   class_names=class_names, 
                   rounded=True)
    plt.show()
# Example usage (assuming you have a trained classifier)
# classifier = tree.DecisionTreeClassifier()
# classifier.fit(X_train, y_train) 
plot_decision_tree(classifier, ['Age','EstimatedSalary'], ['Not Purchased', 'Purchased'])
```

```python
importances = classifier.feature_importances_
print("Feature importances:")
for name, importance in zip(['Age', 'EstimatedSalary'], importances):
    print(f"{name}: {importance:.4f}")
```

```python
import graphviz
# DOT data
dot_data = tree.export_graphviz(clf2, out_file=None,
                                feature_names=['Age','EstimatedSalary'],
                                class_names=['Not Purchased', 'Purchased'],
                                filled=True)
# Draw graph
graph = graphviz.Source(dot_data, format="png")
graph
```

```python
#pip install graphviz
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz 
iris = load_iris()
df_iris = pd.DataFrame(iris['data'], 
columns=iris['feature_names'])
df_iris['target'] = iris['target']
df_iris.head()
# Train the Decision tree model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)
dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=iris.feature_names,  
                                class_names=iris.target_names,  
                                filled=True, rounded=True,  
                                special_characters=True)
graph = graphviz.Source(dot_data)  
graph
```
